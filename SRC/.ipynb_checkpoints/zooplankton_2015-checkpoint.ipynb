```python
# imporing necessary packages
import pandas as pd
import numpy as np
import sympy.mpmath as math
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import matplotlib.patches as mpatches
import pylab
%matplotlib inline
```

```python
def data_load(filename):
    '''Loads data in CSV format, produces pandas dataframe for data, 
    and converts index to datetime, only input required is 
     the file paths'''
    ''' input must have date column labeled 'DATE', and the file path 
    must be in quotes'''
    
    assert type(filename) == str, 'filename_1 should be a string'

    # loading the first dataset and changing the index column to the date
    dataset = pd.read_csv(filename, sep=',', index_col='DATE')
    # converting the date index to datetime
    dataset.index = pd.to_datetime(dataset.index)
    
    return dataset
    
```

```python
# plotting the zooplankton abundance in the Discovery Passage over time 

# setting the figure size
fig = plt.figure(figsize=(12, 5))

# adding a title to the graph
plt.suptitle('Zooplankton Abundance in the Discovery Passage', fontsize=24, y=1.05)

# plotting the zooplankton over time
plt.plot(zoo_2015.index, zoo_2015.Total, 'maroon', linewidth=2.2)

# labeling the x and y axes
plt.xlabel('Date', fontsize=16, )
plt.ylabel('Zooplankton Abundance', fontsize=16)

# setting the extremities to the y axis
plt.ylim([0, 2500])


```

```python
fig = plt.figure(figsize=(12, 7))

# Adding a title to the graph
plt.suptitle('2015 Zooplankton and Water Temperature', fontsize=30, y=1.05)

# layering the three axes, which contain the three varaibles, on top of one another to create a single subplot
# layering completed with twinx() method
ax1 = plt.subplot()
ax2 = ax1.twinx()
ax3 = ax1.twinx()
ax4 = ax1.twinx()


# plotting the 3 variables into the predetermined plot outlined above
# Setting the color and width of lines on the graph
ax1.plot(zoo_2015.index, zoo_2015.Total, 'maroon', linewidth =3)
ax2.plot(temp_2015.index, temp_2015.m_0, 'sage', linewidth=3)
ax3.plot(temp_2015.index, temp_2015.m_5, 'darkgreen', linewidth=3)
ax4.plot(temp_2015.index, temp_2015.m_10, 'darkslategray', linewidth=3)
    
# moving the spine that the zooplankton was plotted onto towards the right so it isnt on top of the second spine(ax2)
temp_2015.m_0_spine = ax3.spines['right']
temp_2015.m_0_spine.set_position(('axes', 1.14))

temp_2015.m_10_spine = ax4.spines['right']
temp_2015.m_10_spine.set_position(('axes', 1.24))


# Adjusting the color of the axes ticks to match line color
ax1.xaxis.set_tick_params(labelcolor='k', size=18)
ax1.yaxis.set_tick_params(labelcolor='maroon', size=18)
ax2.yaxis.set_tick_params(labelcolor='sage', size=18)
ax3.yaxis.set_tick_params(labelcolor='darkgreen', size=18)
ax4.yaxis.set_tick_params(labelcolor='darkslategray', size=18)

# Making axes labels, selecting their color, and selecting their size
ax1.set_xlabel('Date', fontsize=20)
ax1.set_ylabel('Zooplankton Abundance' + ' (/m$^3$)', fontsize = 18, color ='maroon')
ax2.set_ylabel('Sea surface Temperature' + ' ($^o$C)', fontsize = 18, color='sage')
ax3.set_ylabel('Water Temperature at 5m' + ' ($^o$C)', fontsize = 18, color='darkgreen')
ax4.set_ylabel('Water Temperature at 10m' + ' ($^o$C)', fontsize = 18, color='darkslategray')

ax2.set_ylim(9,14)
ax3.set_ylim(9,14)

legend = ax2.legend(loc='upper center', shadow=True)

plt.show()
```

```python
zoo_2009 = zoo_09_12['2009']
zoo_2010 = zoo_09_12['2010']
zoo_2011 = zoo_09_12['2011']
zoo_2012 = zoo_09_12['2012']
```

```python
zoo_2015_w = zoo_2015.resample('W', how=('mean'))
```

```python
zoo_2009_w.index = pd.date_range(start='2015-03-01',end='2015-07-05',freq='W')
zoo_2010_w.index = pd.date_range(start='2015-03-07',end='2015-06-28',freq='W')
zoo_2011_w.index = pd.date_range(start='2015-03-08',end='2015-07-04',freq='W')
zoo_2012_w.index = pd.date_range(start='2015-03-29',end='2015-07-02',freq='W')
```

```python
fig = plt.figure(figsize=(12, 7))

# Adding a title to the graph
plt.suptitle('2009-2015 Zooplankton Abundance', fontsize=30, y=1.05)

# layering the three axes, which contain the three varaibles, on top of one another to create a single subplot
# layering completed with twinx() method
ax1 = plt.subplot()
ax2 = ax1.twinx()
ax3 = ax1.twinx()
ax4 = ax1.twinx()
ax5 = ax1.twinx()


# plotting the 3 variables into the predetermined plot outlined above
# Setting the color and width of lines on the graph
ax1.plot(zoo_2015_w.index, zoo_2015_w.Total, 'mediumaquamarine', linewidth =3)
ax2.plot(zoo_2012_w.index, zoo_2012_w.Total_Zooplankton_Abundance, 'sage', linewidth=3)
ax3.plot(zoo_2011_w.index, zoo_2011_w.Total_Zooplankton_Abundance, 'darkgreen', linewidth=3)
ax4.plot(zoo_2010_w.index, zoo_2010_w.Total_Zooplankton_Abundance, 'darkslategray', linewidth=3)
ax5.plot(zoo_2009_w.index, zoo_2009_w.Total_Zooplankton_Abundance, 'midnightblue', linewidth=3)
    
# moving the spine that the zooplankton was plotted onto towards the right so it isnt on top of the second spine(ax2)



# Adjusting the color of the axes ticks to match line color
ax1.xaxis.set_tick_params(labelcolor='k', size=18)
ax1.yaxis.set_tick_params(labelcolor='k', size=18)

ax1.yaxis.set_ticks_position('left')
ax1.xaxis.set_ticks_position('bottom')

# Making axes labels, selecting their color, and selecting their size
ax1.set_xlabel('Date', fontsize=20)
ax1.set_ylabel('Zooplankton Abundance' + ' (/m$^3$)', fontsize = 18, color ='black')

formatter = DateFormatter('%m/%d')
ax1.xaxis.set_major_formatter(formatter)  

ax1.set_ylim(0,8000)
ax2.set_ylim(0,8000)
ax3.set_ylim(0,8000)
ax4.set_ylim(0,8000)
ax5.set_ylim(0,8000)

mediumaquamarine_line = mlines.Line2D([], [], color='mediumaquamarine', marker='', linewidth=3, markersize=20, label='2015')
sage_line = mlines.Line2D([], [], color='sage', marker='', linewidth=3, markersize=20, label='2012')
darkgreen_line = mlines.Line2D([], [], color='darkgreen', marker='', linewidth=3, markersize=20, label='2011')
darkslategray_line = mlines.Line2D([], [], color='darkslategray', marker='', linewidth=3, markersize=20, label='2010')
midnightblue_line = mlines.Line2D([], [], color='midnightblue', marker='', linewidth=3, markersize=20, label='2009')

plt.legend(handles=[mediumaquamarine_line, sage_line, darkgreen_line, darkslategray_line, midnightblue_line])

ax2.yaxis.set_visible(False)
ax3.yaxis.set_visible(False)
ax4.yaxis.set_visible(False)
#ax5.yaxis.set_visible(False)

plt.savefig('../Results/2009-2015_zooplankton_abundance.pdf')

plt.show()
```
